```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl = pd.read_csv('C:\\Users\\tagomuo\\Downloads\\EPL.csv')
epl.head()
```

```python
epl.info()
```

```python
epl.describe()
```

```python
# searching for null values
epl.isna().sum()
```

```python
# create  2 new columns
epl['MinsPerMatch'] = (epl['Mins']/epl['Matches']).astype(int)
epl['GoalsPerMatch'] = (epl['Goals'] / epl['Matches']).astype(float)

epl.head()
```

```python
## Total goals scored
Total_Goals = epl['Goals'].sum()
print(Total_Goals)
```

```python
##  Penalty Goals
PenaltyGoals = epl['Penalty_Goals'].sum()
print(PenaltyGoals)
```

```python
### Penalty attempts

PenaltyAttempts = epl['Penalty_Attempted'].sum()
print(PenaltyAttempts)
```

```python
## Percentage of Penalty converted
PercentagePenaltyConverted = (PenaltyGoals/PenaltyAttempts) * 100
print(PercentagePenaltyConverted)
```

```python
## Pie chart for Penalty Missed vs scored 

plt.figure(figsize=(13,6))
pl_not_scored = epl['Penalty_Attempted'].sum() - PenaltyGoals
data = [pl_not_scored, PenaltyGoals]
labels = ['Penalties Missed', 'Penalty Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%' )
plt.show()
```

```python
###  Unique Positions
epl['Position'].unique()
```

```python
### Total FW Players

epl[epl ['Position'] == 'FW']
```

```python
## Players from different nations

np.size ((epl['Nationality'].unique()))
```

```python
### Most Players from Particaular nations

nationality = epl.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette('magma'))
```

```python
### Clubs with max players in their squad
epl['Club'].value_counts().nlargest(5).plot(kind = 'bar', color=sns.color_palette('viridis'))
```

```python
### Clubs with least players in their squad
epl['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color=sns.color_palette('viridis'))
```

```python
## Players based on age group

under20 = epl[epl['Age'] <= 20 ]
age20_25 = epl[(epl['Age'] > 20 ) & (epl['Age'] <= 25) ]
age25_30 = epl[(epl['Age'] > 25 ) & (epl['Age'] <= 30 )]
above30 = epl[epl['Age'] > 30 ]
               
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
mylabels = ['<=20', '>20 & <=25', '>25 & <=30', '>30']
plt.title('Total Players with Age Groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct='%.1f%%')
plt.show()
```

```python
### Total under 20 Players in each club

players_under_20 = epl[epl['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind = 'bar', color =sns.color_palette('cubehelix'))
```

```python
## Under 20 Players in Manchester United

players_under_20[players_under_20["Club"] == 'Manchester United']
```

```python
### Under 20 Players in Chelsea

players_under_20[players_under_20["Club"] == 'Chelsea']
```

```python
## Average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x= 'Club', y = 'Age', data = epl)
plt.xticks(rotation = 90)
```

```python
num_player = epl.groupby('Club').size()
data = (epl.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
## Total Assists from each club

Assists_by_clubs = pd.DataFrame(epl.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=Assists_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists', fontsize=30)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs vs Total Assists', fontsize = 20)
```

```python
## Top 10 Assists

top_10_assists = epl[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns = 'Assists')
top_10_assists
```

```python
Goals_by_clubs = pd.DataFrame(epl.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y ='Goals', data=Goals_by_clubs.sort_values(by='Goals'), palette = 'rocket')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Goals', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize = 20)
```

```python
## Most goals by Player
top_10_goals = epl[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns='Goals')
top_10_goals
```

```python
## Goals Per match

top_10_goals_per_match = epl[['Name', 'GoalsPerMatch', 'Matches', 'Goals']].nlargest(n=10, columns = 'GoalsPerMatch')
top_10_goals_per_match
```

```python
## Pie chart - Goals with assist and without assist

plt.figure(figsize = (14,7))
assists = epl['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```
